```python
import pandas as pd
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from scipy import stats
from datetime import datetime
```

```python
data_path_in = Path('../data/raw/dummy_RAW_INPUT_METRICS.csv')
df_in = pd.read_csv(data_path_in)

data_path_or = Path('../data/raw/dummy_RAW_ORDERS.csv')
df_or = pd.read_csv(data_path_or)
```

```python
# ============================================================================
# Data Preparation - Create Working Copy
# ============================================================================
# Create a copy of both data_frames for analysis
dfin = df_in.copy()
dfor = df_or.copy()

print("=" * 80)
print("METRICS DATAFRAME")
print("=" * 80)
print(f"\nOriginal shape: {dfin.shape}")
print(f"\nInitial data types:\n{dfin.dtypes}")
print(f"\nInitial null values:\n{dfin.isnull().sum()}")
print("=" * 80)
print("ORDERS DATAFRAME")
print("=" * 80)
print(f"\nOriginal shape: {dfor.shape}")
print(f"\nInitial data types:\n{dfor.dtypes}")
print(f"\nInitial null values:\n{dfor.isnull().sum()}")
```

```python
# ============================================================================
# removing null values
# ============================================================================
dfin = dfin.fillna(0, inplace=True)
dfor = dfor.fillna(0, inplace=True)
```

```python
# ============================================================================
# NaN and Null values removed
# ============================================================================


print("=" * 80)
print("METRICS DATAFRAME")
print("=" * 80)
print(f"\nOriginal shape: {dfin.shape}")
print(f"\nInitial data types:\n{dfin.dtypes}")
print(f"\nInitial null values:\n{dfin.isnull().sum()}")
print("=" * 80)
print("ORDERS DATAFRAME")
print("=" * 80)
print(f"\nOriginal shape: {dfor.shape}")
print(f"\nInitial data types:\n{dfor.dtypes}")
print(f"\nInitial null values:\n{dfor.isnull().sum()}")
```

```python
# ============================================================================
# Making a reshape of both data frames columns
# ============================================================================
dfin = dfin.melt(
    id_vars=['COUNTRY', 'CITY', 'ZONE','ZONE_TYPE','ZONE_PRIORITIZATION','METRIC'],
    value_vars=['L8W_ROLL', 'L7W_ROLL', 'L6W_ROLL','L5W_ROLL','L4W_ROLL','L3W_ROLL','L2W_ROLL','L1W_ROLL','L0W_ROLL'],
    var_name='NW_AGO',
    value_name='MVALUE'
)

dfor = dfor.melt(
    id_vars=['COUNTRY', 'CITY', 'ZONE','METRIC'],
    value_vars=['L8W', 'L7W', 'L6W','L5W','L4W','L3W','L2W','L1W','L0W'],
    var_name='NW_AGO',
    value_name='ORDERS'
)

dfin['NW_AGO'] = dfin['NW_AGO'].str.extract(r'(\d+)').astype(int)
dfor['NW_AGO'] = dfor['NW_AGO'].str.extract(r'(\d+)').astype(int)


#Deleting METRIC Column on Orders Dataframe
if 'METRIC' in dfor.columns:
    dfor = dfor.drop('METRIC', axis=1)
    print("\n'METRIC' removed from dataframe")
```

```python
# ============================================================================
# PROCESSED DATAFRAMES BACKUP
# ============================================================================

output_dir = Path('../data/processed')
output_dir.mkdir(parents=True, exist_ok=True)

file_path1 = output_dir / 'input_metrics_processed.csv'
file_path2 = output_dir / 'orders_metrics_processed.csv'

dfin.to_csv(file_path1, index=False)
dfor.to_csv(file_path2, index=False)
```
